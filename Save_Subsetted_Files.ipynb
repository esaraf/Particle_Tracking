```python
import xarray as xr
import glob
import numpy as np
```

```python
# -----------------------------------------------------------------------------
# Input variables, change these
# -----------------------------------------------------------------------------

# years downloaded
yr_wnt = 2002

# days downloaded
day_bgn = 1
day_end = 365

# directory path of netcdf files
# dir_in = '/Users/esaraf/Desktop/NOAA/GlobCurrent/GlobCurrent.geostrophic.data/geostrophic/'
# dir_in = '/home/isaac/eftp.ifremer.fr/data/globcurrent/v3.0/global_025_deg/geostrophic/'
dir_in = '/Users/esaraf/Desktop/NOAA/Particles_Total_Velocities/2002data/globcurrent/v3.0/global_025_deg/total_hs/{}/'.format(yr_wnt)


# Subset data to get CCS
lon_bgn = -150
lon_end = -110
lat_bgn = 20
lat_end = 50

# contour this variable
var_wnt = ['eastward_eulerian_current_velocity',
           'northward_eulerian_current_velocity']

# output directory, you will need to create this directoy on your desktop
dir_out = '/Users/esaraf/Desktop/NOAA/Particles_ccs_files_total/2002'.format(yr_wnt)

# -----------------------------------------------------------------------------
# END: Input variables, change these
# -----------------------------------------------------------------------------
```

```python
# size of input variables
num_var_wnt = len(var_wnt)

# construct day array
days_wnt = np.arange(day_bgn, day_end+1)
num_days_wnt = len(days_wnt)
```

```python
# loop all days, save files ot output directory
for i in range(num_days_wnt):
    print('yr={}, day={:03d}'.format(yr_wnt, days_wnt[i]))
        
    # construct directory that has year and day subdirectories
    dir_yr_day = '{}/{}/{:03d}'.format(dir_in, yr_wnt, days_wnt[i])

    # list the netcdf files in this directory
    fn_check = '{}/*.nc'.format(dir_yr_day)
    fn_nc = glob.glob(fn_check)

    # check to see if one file is in the directory
    num_fn_nc = len(fn_nc)

    if num_fn_nc == 1:
        # open netcdf file using Xarray package (similar to Pandas but for geophysical data)
        ds1 = xr.open_dataset(fn_nc[0])

        # variables and coords
        var_ds1 = list(ds1.keys())
        coord_ds1 = list(ds1.coords.keys())

        # from the dataset create u and v DataArrays
        da_u = ds1[var_wnt[0]]
        da_v = ds1[var_wnt[1]]

        # subset u and v for CCS
        # this is where we use xarray sel function to subset the
        # global data down to the CCS region
        da_u_ccs = da_u.sel(lon=slice(lon_bgn, lon_end)).sel(
            lat=slice(lat_bgn, lat_end))

        da_v_ccs = da_v.sel(lon=slice(lon_bgn, lon_end)).sel(
            lat=slice(lat_bgn, lat_end))

        # create xarray U and V dataset from the u and v DataArrays
        ds1_out = da_u_ccs.to_dataset(name='U')
        ds1_out['V'] = da_v_ccs

        # write U dataset to a netcdf file
        fn_out = '{}/{}_{:03d}_total_uv.nc'.format(dir_out, yr_wnt, days_wnt[i])
        ds1_out.to_netcdf(fn_out)
```
